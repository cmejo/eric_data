## imputation missing 2016 for recycling and landfill

```python
selected_by_sites = selected.set_index(['site_code','year']).sort_index()
```

```python
imputed_recycling = selected_by_sites.groupby('site_code')['waste_recycling_tonnes'].apply(lambda group: group.interpolate())
```

```python
selected_by_sites['waste_recycling_tonnes'] = imputed_recycling
```

```python
filled_recycling = selected_by_sites.groupby('site_code')['waste_recycling_tonnes'].bfill().ffill()
```

```python
selected_by_sites['waste_recycling_tonnes'] = filled_recycling
```

```python
imputed_landfill = selected_by_sites.groupby('site_code')['waste_landfill_tonnes'].apply(lambda group: group.interpolate())
```

```python
selected_by_sites['waste_landfill_tonnes'] = imputed_landfill
```

```python
selected_by_sites.info(verbose=True)
```

## Generate aggregate values

```python
selected_by_sites['waste_bio_tonnes'] = selected_by_sites['waste_incineration_tonnes'] + selected_by_sites['waste_alternative_tonnes']
```

```python
selected_by_sites['waste_domestic_tonnes'] = selected_by_sites['waste_landfill_tonnes'] + selected_by_sites['waste_recycling_tonnes']
```

```python
selected_by_sites['waste_total_tonnes'] = selected_by_sites['waste_incineration_tonnes'] + selected_by_sites['waste_alternative_tonnes'] + selected_by_sites['waste_landfill_tonnes'] + selected_by_sites['waste_recycling_tonnes']
```

```python
selected_by_sites['occupied_area_m2'] = selected_by_sites['area_patient_m2'] + selected_by_sites['area_nonpatient_m2']
```

```python
##generate static indices
## 1) all types of waste / (area_patient and/or area_nonpatient)
## 1a) waste_bio / (patient area)
## 1b) waste_domestic / (area_nonpatient)
## 2) electricity consumed / (area_patient + area_nonpatient)
## 3) gas_consumed / (volume_heated_m3)
## 4) green index
```

```python
selected_by_sites['waste_total_index'] = selected_by_sites['waste_total_tonnes']/selected_by_sites['occupied_area_m2']
```

```python
selected_by_sites['waste_incineration_index'] = selected_by_sites['waste_incineration_tonnes']/selected_by_sites['area_patient_m2']
selected_by_sites['waste_alternative_index'] = selected_by_sites['waste_alternative_tonnes']/selected_by_sites['area_patient_m2']
selected_by_sites['waste_bio_index'] = selected_by_sites['waste_bio_tonnes']/selected_by_sites['area_patient_m2']
```

```python
selected_by_sites['waste_landfill_index'] = selected_by_sites['waste_landfill_tonnes']/selected_by_sites['occupied_area_m2']
selected_by_sites['waste_recycling_index'] = selected_by_sites['waste_recycling_tonnes']/selected_by_sites['occupied_area_m2']
selected_by_sites['waste_domestic_index'] = selected_by_sites['waste_domestic_tonnes']/selected_by_sites['occupied_area_m2'] 
```

```python
selected_by_sites['electricity_index'] = selected_by_sites['consumed_electricity_kwh']/selected_by_sites['occupied_area_m2']
```

```python
selected_by_sites['gas_index_vol'] = selected_by_sites['consumed_gas_kwh']/selected_by_sites['volume_heated_m3']
```

```python
selected_by_sites['electricity_green_index'] = selected_by_sites['consumed_green_kwh']/selected_by_sites['occupied_area_m2']
```
